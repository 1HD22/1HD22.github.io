```python
import json

with open("../data/Output.json", "w") as text_file:
    text_file.write(json.dumps([{"className": i, "axes": [{"axis":j, "value":k} for (j,k) in zip(data_numer.columns, data_numer.loc[i, :])]} for i in data_numer.index]))
```
