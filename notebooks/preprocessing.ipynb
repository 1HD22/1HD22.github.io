```python
import pandas as pd
```

```python
data = pd.read_csv("../data/Self-Introduction.csv").drop(["Timestamp", "Unnamed: 23", "avg"], axis=1).drop([43, 44]).set_index("ALIAS")
data
```

```python
cols = data.columns
names = data.index
```

```python
cols
```

```python
# unique values for first col

unique_Orig = [data.iloc[:, i].unique() for i in range(len(cols))]
```

```python
unique_Orig[0]
```

```python
data.iloc[:, 0] = (data.iloc[:, 0]
                   .fillna("KTH")
                   .replace(r".*KTH.*", "KTH", regex=True)
                   .replace("The Hong Kong University of Science and Technology", "HKUST (Hong Kong)"))
data.iloc[:, 0].unique()
```

```python
unique_Orig[1]
```

```python
data.iloc[:, 1] = (data.iloc[:, 1]
                   .replace("18", "2018")
                   .replace('Masters programme in October 2022 (BSc in 2018)', "2022")
                   .replace('First Year of the master programme', "2023"))
data.iloc[:, 1].unique()
```

```python
unique_Orig[2].__len__()
```

```python
data.iloc[:, 2] = pd.to_datetime(data.iloc[:,2]).dt.strftime("%m/%Y")
data.iloc[:, 2]
```

```python
unique_Orig[3]
```

```python
data.iloc[:, 3] = (data.iloc[:, 3]
                   .str.title()
                   .str.strip()
                   .replace("Data Science And Technology", "Data Science"))
data.iloc[:, 3]
```

```python
data.iloc[:, 3].unique()
```

```python
unique_Orig[4]
```

```python
unique_Orig[5].__len__()
```

```python
data.iloc[:, 5] = data.iloc[:, 5].apply(lambda x: x.replace("cicling", "cycling").replace("theatre", "theater") if not isinstance(x, float) else x)
```
